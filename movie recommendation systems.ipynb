```python
import numpy as np
from sklearn.metrics import accuracy_score, recall_score
ground_truth = np.array([4, 5, 3, 2, 1, 4, 5, 2, 3, 1])
predicted_ratings = np.array([3.8, 4.5, 3.2, 2.1, 1.5, 4.0, 4.8, 2.5, 3.1, 1.2])
threshold = 3.5
ground_truth_binary = (ground_truth >= threshold).astype(int)
predicted_binary = (predicted_ratings >= threshold).astype(int)
accuracy = accuracy_score(ground_truth_binary, predicted_binary)
print(f'Accuracy: {accuracy:.2f}')
recall = recall_score(ground_truth_binary, predicted_binary)
print(f'Recall: {recall:.2f}')
```
